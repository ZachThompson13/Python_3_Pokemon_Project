Okay, so we know that certain types of attacks do half, standard, or double damage, but how do we actually calculate the damage? Well, damage depends partly on how much power each move has. Below is the power of each type of attack:

How is damage actually calculated? According to this pretty complicated formula from this website https://bulbapedia.bulbagarden.net/wiki/Damage#Damage_calculation :

$$Damage=(\frac{(\frac{2 \times Level}{5}+2) \times Power \times \frac{A}{D}}{50}) \times Modifier$$

Let's simplify this a bit by assuming that all Pokémon are on Level 50 temporarily:

$$Damage=(\frac{22 \times Power \times \frac{A}{D}}{50}) \times Modifier$$

What are A and D? A is the effective Attack stat of the attacking Pokémon and D is the is the effective Defense stat of the defending Pokémon. This info for each Pokémon is listed below:

Actually, let's put all Pokémon data into one dictionary for simplicity:

What is the modifier in the damage formula above? Our simplified modifier will be:

$$ Modifier = Critical \times Random \times Type $$

**Random** is a random integer percentage between 0.85 and 1.00 (inclusive).

Recall that **Type** can be 0.5 (not very effective), 1 (normally effective), or 2 (super effective), depending on how effective the move's type is against the defending Pokémon's type.

Actually, the type can also take on the values ¼ and 4. For targets that have two types (every Pokémon is either one or two types ), such as Bulbasaur, who is both a Grass and Poison type, overall type effectiveness is the product of the effectiveness against each of its types:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is super effective against both of the opponent's types (such as Dig, a Ground-type move, used against an Aggron, a Steel/Rock Pokémon), then the move does 2x2=4 times the damage;

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is not very effective against both of the opponent's types (such as Wake-Up Slap, a Fighting-type move, used against a Sigilyph, a Psychic/Flying Pokémon), then the move only does 1/2x1/2=¼ of the damage;

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is super effective against one of the opponent's types but not very effective against the other (such as Razor Leaf, a Grass-type move, used against a Gyarados, a Water/Flying Pokémon), then the move deals normal damage 2x1/2=1;

**Critical** is 2 if you have landed a critical hit and 1 otherwise. What is a critical hit? This is of the one last things to understand. A critical hit is a random event that multiplies the damage of a damage-dealing move. When one occurs, the game will display the message "A critical hit!" after the damage has been dealt. 

How does one calculate the probability that the attack was critical? The probability that an attacking Pokémon landed a critical hit based on that attacking Pokémon's speed is: $$ P = Speed / 512 $$

Thus, one can generate a random  number between 0 and 511 and if that number is less than the attacking Pokémon's speed then Critical=2.

Therefore, there will be TWO different random components to our damage calculation, represented by Random and Critical in the formulas above, and we'll need to add another attribute, speed, to our Pokémon dictionary:


Now let's return to levels. When a Pokémon beats another Pokémon, he gains their experience points. With enough experience points, the Pokémon achieves the next level. Levels go from 1 to 100. When a player just starts out, they have 0 experience points and are on level 1. Once they gain experience points, the function that models experience points versus level is the cube root function:

$$ Level = (Experience)^{1/3} $$

We can now take the level of the attacker into account when calculating damage:

$$Damage=(\frac{(\frac{2 \times Level}{5}+2) \times Power \times \frac{A}{D}}{50}) \times Modifier$$

````python
class Move():
    """Stores info about a move/attack"""
    def __init__(self, name:str, power:int, type:str,
                 super_effective_against:list[str], not_very_effective_against:list[str]):
        
        self.name = name
        self.power = power
        self.type = type
        self.super_effective_against = super_effective_against
        self.not_very_effective_against = not_very_effective_against

    def __repr__(self):
        return f'{self.name}: {self.power}'
    def __str__(self):
        return self.name

class Pokemon():
    """Stores info about a pokemon"""
    def __init__(self, Name:str, Type:list[str], HP:int, Moves:list[Move], 
                 Attack:int, Defense:int, Speed:int, Experience:int, level=1):
        
        self.name = Name
        self.type = Type
        self.max_HP = self.HP = HP
        self.attack = Attack
        self.defence = Defense
        self.speed = Speed
        self.exp_given = Experience
        self.current_exp = 0
        self.level = level
        self.caught = False

        self.attacking = False
        self.switching = False
        self.using_item = False
        self.fainted = False

        self.moves = [Move(**MOVES_DICTIONARY[move]) for move in Moves]

    @property
    def next_level_exp(self):
        return self.level ** 3
    
    def __lt__(self, other):
        return self.speed > other.speed

    def update_level(self, exp):
        """Updates a pokemon's level given the amount of exp"""
        self.current_exp += exp

        while self.current_exp >= self.next_level_exp:
            self.current_exp -= self.next_level_exp
            self.level += 1
            print(f'Level up! {self.name} is now level {self.level}.')
        

    def __repr__(self):
        return f'{self.name} level {self.level}: {self.HP}/{self.max_HP}'
    def __str__(self):
        return f'{self.name}'

def select_move(moves:list[Move]) -> Move:
        """Selects a move"""
        print('\nYour moves:')
        for move in moves:
            print(move)
        print()
        
        while True:
                choice = input('Select an attack: ').title()
                for move in moves:
                    if choice == move.name:
                        return move
                    
                else:
                    print('Invalid Choice')
                    continue
                
def select_target(targets: list[Pokemon]) -> Pokemon:
        """
        Selects a target for an action
        Returns the target pokemon
        """

        print('Your choices are:')
        for target in targets:
            print(target)
        print()
        
        while True:
            choice = input('Select a target: ').title()
            for pokemon in targets:
                if choice == pokemon.name and not pokemon.fainted:
                    return pokemon
                
            else:
                print('Invalid Choice')
                continue


class Team():
    def __init__(self, *pokemon:Pokemon, team_name:str='Wild Pokemon'):
        if len(pokemon) > 6:
            raise ValueError
        
        self.name = team_name
        if team_name == 'Wild Pokemon':
            self.full_team = list(pokemon)
        else:
            self.full_team = list(pokemon) + (6 - len(pokemon)) * [None] # creates empty placeholders
            
        self.active_pokemon:list[Pokemon] = []
        self.benched_pokemon = list(pokemon)
        self.fainted_pokemon:list[Pokemon] = []
        
        self.total_exp_given = sum([p.exp_given for p in self.full_team if p])


    def __str__(self):
        return self.name
    
    def __eq__(self, other):
        return self is other
    
    def remove_pokemon(self, pokemon:Pokemon, position) -> None:
        """Removes a pokemon from a team"""
        self.full_team[position] = None
        if pokemon in self.benched_pokemon:
            self.benched_pokemon.remove(pokemon)

    def add_pokemon(self, pokemon:Pokemon, posisiton) -> None:
        """Adds a pokemon to a team if under 6 pokemon"""
        self.full_team[posisiton] = pokemon
        self.benched_pokemon += [pokemon]


    def switch_pokemon(self, retreating_pokemon:Pokemon, new_pokemon) -> None:
        """Switchs an active pokemon with one on the bench"""
        if not self.benched_pokemon:
            print('You have no pokemon to switch with!\n')
            retreating_pokemon.switching = False
            return

        if retreating_pokemon in self.active_pokemon: # in case fainting
            self.active_pokemon.remove(retreating_pokemon)

        if not retreating_pokemon.fainted:
            self.benched_pokemon.append(retreating_pokemon)
        if new_pokemon in self.benched_pokemon:
            self.benched_pokemon.remove(new_pokemon)

        self.active_pokemon.append(new_pokemon)
        new_pokemon.attacking = False
        # new pokemon doesn't get their turn this round, so at end of order
        print(f'\n{self.name} recalled {retreating_pokemon}.')
        print(f'"Come back, {retreating_pokemon}!"')
        print(f'{self.name} sent in {new_pokemon}!')
        print(f'"Go {new_pokemon}!"\n')

        return
        
    def resolve_faint(self, pokemon:Pokemon) -> None:
        """Moves a pokemon to fainted"""
        if pokemon in self.active_pokemon:
            self.active_pokemon.remove(pokemon)
        self.fainted_pokemon.append(pokemon)
        
        print()
        if self.benched_pokemon:
            self.switch_pokemon(pokemon, select_target(self.benched_pokemon))

        return

class Pokedex(list):
    """Stores which pokemon a player has caught"""
    def __init__(self):
        self.collected_pokemon:list[Pokemon] = []
        self.collected_pokemon_names:list[str] = []
    
    def __len__(self):
        return len(self.collected_pokemon)

    def __str__(self):
        return '\n'.join([str(p) for p in self.collected_pokemon])
    
    def add_pokemon(self, pokemon:Pokemon) -> None:
        """Adds a pokemon to the pokedex"""
        # Adding new pokemon
        if pokemon.name not in self.collected_pokemon_names:
            pokemon.HP = pokemon.max_HP
            pokemon.fainted = False
            self.collected_pokemon += [pokemon]
            self.collected_pokemon_names += [pokemon.name]

        # if pokemon already caught, heal and level up duplicate in your team
        else:
            for duplicate in self.collected_pokemon:
                if pokemon.name == duplicate.name:
                    duplicate.fainted = False
                    duplicate.HP = duplicate.max_HP
                    duplicate.update_level(duplicate.next_level_exp - duplicate.current_exp)

        return
        
   
    def update(self) -> None:
        """Prints a 'pokedex updated' message"""
        print('\nYour Pokedex was updated!')
        print('It now contains:')
        print(self)
        print()
        return

class Action():
    """stores a pokemon's intended action"""
    def __init__(self, pokemon:Pokemon, move, new_pokemon, item, team:Team):
        self.pokemon = pokemon
        self.move = move
        self.new_pokemon = new_pokemon
        self.item = item
        self.team = team

class Pokemon_Battle():
    def __init__(self, team1:Team, team2:Team, pokedex:Pokedex, doubles:bool=False):
        self.team1 = team1
        self.team2 = team2
        self.teams = (team1, team2)
        self.doubles = doubles
        self.turn = 0
        self.pokedex = pokedex

        # resetting active and benched
        for team in self.teams:
            team.active_pokemon = []
            team.benched_pokemon = [p for p in team.full_team if p]
            team.fainted_pokemon = []
            team.active_pokemon += [team.benched_pokemon[0]]
            team.benched_pokemon.pop(0)
            if doubles:
                team.active_pokemon += [team.benched_pokemon[1]]
                team.benched_pokemon.pop(0)

    def cause_move(self, attacker:Pokemon, defender:Pokemon, move:Move) -> None:
        """
        Executes a move given the attacker, defender, and the move.
        Prints outcome of move and if defender fainted.
        If defender fainted, sets defender.fainted to True.
        Returns Nothing.
        """
        
        print(f'\n{attacker} used {move}!')
        if move not in attacker.moves:
            print(f'But it failed!')

        # Effectiveness calculator and messages
        effectiveness = 1
        for type in defender.type:
            if type in move.super_effective_against:
                effectiveness *= 2
            elif type in move.not_very_effective_against:
                effectiveness *= 0.5
        if effectiveness > 1:
            print('Super Effective!')
        elif 0 < effectiveness < 1:
            print('Not very effective...')
        elif effectiveness == 0:
            print('No Effect.')
        
        # Random aspect
        critical = 1
        if random.randint(0,511) < attacker.speed:
            critical = 2
            print('Critical Hit!')
        r = random.randint(85,100) * 0.01

        modifier = effectiveness * critical * r

        # damage calculation
        damage = int((((0.4 * attacker.level + 2) * move.power * (attacker.attack/defender.defence)) / 50) * modifier)

        # Deals with excess damage and fainting
        if damage >= defender.HP:
            defender.HP = 0
            print(f'{defender} fainted!')
            defender.fainted = True
        else:
            defender.HP -= damage
        
        print(f'{defender}: {defender.HP}/{defender.max_HP}')
        return
    
    def use_item(self, target:Pokemon, item) -> None:
        """Uses a given item"""
        items = []
        pass

    def individual_select_action(self, pokemon, action:str) -> None:
        """Chooses to attack, switch, or use an item for an individual pokemon."""            
        if action == 'attack':
            pokemon.attacking = True
            pokemon.switching = False
            pokemon.using_item = False
            
        elif action == 'switch':
            pokemon.attacking = False
            pokemon.switching = True
            pokemon.using_item = False
            
        elif action == 'item':
            pokemon.attacking = False
            pokemon.switching = False
            pokemon.using_item = True

        return None

    def choose_actions(self) -> tuple[list[Action], list[Action]]:
        """
        Chooses actions for all pokemon based on user input
        Returns tuple of all pokemon attacking and all pokemon switching.
        """
        # choosing what to do this turn; choices happen before attacks
        attacks:list[Action] = []
        switches:list[Action] = []

        # player team choices
        for pokemon in self.team1.active_pokemon:
            print(f"\n{pokemon}'s turn!\n")
            print('Would you like to attack or switch?')
            
            actions = ('attack', 'switch')
            action = None
            while action not in actions:
                action = input('Attack or switch?').lower()
                if action == 'switch' and not self.team1.benched_pokemon:
                    print("You don't have any pokemon to switch with!")
                    continue

            self.individual_select_action(pokemon, action)
            if pokemon.switching and self.team1.benched_pokemon:
                print('Which pokemon will you send in?')
                new_pokemon = select_target(self.team1.benched_pokemon)
                move = None
                item = None
            elif pokemon.attacking:
                print('Select a move.')
                new_pokemon = None
                move = select_move(pokemon.moves)
                item = None
            else:
                new_pokemon = None
                move = None
                item = None

            if move: 
                attacks += [Action(pokemon, move, new_pokemon, item, self.team1)]
            elif new_pokemon:
                switches += [Action(pokemon, move, new_pokemon, item, self.team1)]
        
        # Computer always attacks
        for pokemon in self.team2.active_pokemon:
            self.individual_select_action(pokemon, 'attack')
            move = random.choice(pokemon.moves)

            attacks += [Action(pokemon, move, None, None, self.team2)]


        return (switches, attacks)

    def execute_attack(self, active_pokemon:Pokemon, initial_pokemon:list[Pokemon], move:Move) -> None:
        """Selects a target and executes the attack"""
        # choosing target
        if self.doubles:
            target = select_target(initial_pokemon) # choose target for doubles
        else:
            if active_pokemon in self.team1.active_pokemon:
                target = self.team2.active_pokemon[0] # default target other pokemon in singles
            else:
                target = self.team1.active_pokemon[0]

        # executing move
        if target in self.team1.active_pokemon or target in self.team2.active_pokemon:
            self.cause_move(active_pokemon, target, move)
        else:
            print(f'\n{active_pokemon} used {move}!')
            print("But it missed!")  

        active_pokemon.attacking = False     

        return

    def display_pokemon(self, pokemon:list[Pokemon]) -> None:
        """Prints given pokemon"""
        for p in pokemon:
            print(f'{p}: {p.HP}/{p.max_HP} HP')
        print()
        return

    def check_for_fained(self) -> None:
        """Checks if any pokemon have fainted and resolves them"""
        for team in self.teams:
            for pokemon in team.active_pokemon[:]: # interate thru copy to avoid removing errors
                if pokemon.fainted and not team.benched_pokemon and team.name != 'Wild Pokemon':
                    team.resolve_faint(pokemon)
                    print(f"All of {team.name}'s pokemon have fainted!")
                elif pokemon.fainted:
                    team.resolve_faint(pokemon) # removes from active
        return None

    def battle_round(self) -> bool:
        """
        Runs one round of battle.
        Returns False if all of one team's pokemon have fainted
        Else: returns True
        """
        # Using speed to determine order
        initial_pokemon = self.team1.active_pokemon + self.team2.active_pokemon
        
        # displaying pokemon
        self.display_pokemon(initial_pokemon)    

        # prepares actions
        switches, attacks = self.choose_actions()     
        attacks = sorted(attacks, key=lambda x: x.pokemon) # ordering by speed

        # executing switches before attacks
        for switch in switches:
            team = switch.team
            team.switch_pokemon(switch.pokemon, switch.new_pokemon)

        #executing attacks
        for attack in attacks:
            active_pokemon = attack.pokemon
            if active_pokemon.attacking and not active_pokemon.fainted and active_pokemon in initial_pokemon:
                self.execute_attack(active_pokemon, initial_pokemon, attack.move)

        # deals with fainting and sending more pokemon in
        self.check_for_fained()
        self.turn += 1
        print('\n')

        # checking if battle should continue
        not_none_full_team1 = [p for p in self.team1.full_team if p]
        if (len(self.team1.fainted_pokemon) != len(not_none_full_team1)) and (len(self.team2.fainted_pokemon) != len(self.team2.full_team)):
            return True
        return False

    def post_battle(self) -> Team:
        """
        Gives exp after winning a battle to the active pokemon.
        Returns winning team
        """
        if self.team2.name == 'Wild Pokemon':
            for pokemon in self.team2.full_team:
                if pokemon:
                    print(f'Congratulations! You caught a wild {pokemon}!')
                    pokemon.fainted = False
                    pokemon.attacking = False
                    self.pokedex.add_pokemon(pokemon)
                    self.pokedex.update()
        
        for pokemon in self.team1.active_pokemon:
            exp_gained = self.team2.total_exp_given
            pokemon.update_level(exp_gained)
            print(f'{pokemon} gained {exp_gained} exp! {pokemon.current_exp}/{pokemon.next_level_exp}')
    
        if self.team2.fainted_pokemon == self.team2.full_team:
            winner = self.team1
        else:
            winner = self.team2

        return winner

    def main(self) -> Team:
        keep_battling = True
        while keep_battling:
            keep_battling = self.battle_round()
        
        winner = self.post_battle()
        return winner
````

````python
def trigger_battle(team:Team, pokedex:Pokedex) -> Team:
    """
    Triggers a pokemon battle with a random wild pokemon
    returns the winner
    """
    wild_pokemon = random.choice(list(ALL_POKEMON.keys()))
    wild_pokemon = Pokemon(wild_pokemon, **ALL_POKEMON[wild_pokemon])
    
    print(f'\n\nA wild {wild_pokemon} appeared!')
    if __name__ == '__main__':
        battle = Pokemon_Battle(team, Team(wild_pokemon), pokedex)
        return battle.main()

def change_team(team:Team, pokedex:Pokedex) -> None:
    """Allows a player to change their team order"""
    if len(pokedex) == 1:
        print('You only have 1 pokemon!')
        return
    
    print(f'\nYour current order is:')
    for p in team.full_team:
        print(p)
    print()

    position = 0

    while int(position) < 1 or int(position) > 6:
        print('Which position would you like to change?')

        position = input('Enter position')
        try: # update for decimal input later
            position = int(position) - 1
        except ValueError:
            continue
        
        old_pokemon = team.full_team[position]

        # empty position
        yes, no = ['yes',  'y', 'empty', 'none'], ['n', 'no']
        removed = False
        if old_pokemon:
            while True:
                print('Would you like to empty the position?')
                choice = input('Empty position?').lower()
                if choice in yes:
                    team.remove_pokemon(old_pokemon, position)
                    removed = True
                    break
                if choice in no:
                    break

        # Not forced to add new pokemon after removing one
        if removed:
            print(f'Removed {old_pokemon}!')
            return

        # adding new pokemon
        print('\nWhich pokemon would you like to put there? Choose a pokemon in your pokedex:')
        new_pokemon = None

        availible_pokemon = [p for p in pokedex.collected_pokemon if p not in team.full_team]

        if not availible_pokemon: 
            print('No availible pokemon!')
            return None
        
        while new_pokemon is None:
            new_pokemon = select_target(availible_pokemon)

        if old_pokemon:
            team.remove_pokemon(old_pokemon, position)
        if new_pokemon:
            team.add_pokemon(new_pokemon, position)

    return
````

````python
def get_player_name() -> str:
    """Lets the player input their name and returns it"""
    print("What's your name?")
    name = input('Enter name')
    return name

def choose_starter() -> Pokemon:
    """Lets the player choose a starter from the list"""
    starters = ['Bulbasaur', 'Charmander', 'Squirtle']

    #choosing starter
    choice = None
    while choice not in starters:
        choice = input('Choose starter pokemon').title()

    starter_pokemon = Pokemon(choice, **ALL_POKEMON[choice])
    starter_pokemon.caught = True
    print(f'Congratulations! You have chosen {starter_pokemon}!')
    return starter_pokemon

def set_up_game(name:str, player_pokedex:Pokedex) -> Team:
    """Sets up the game with the player team, pokedex, and some messages"""
    print(f'Hello {name}, and welcome to the wonderful world of pokemon!')
    print('You must choose a starter, either Bulbasaur, Charmander, or Squirtle.')
    
    starter_pokemon = choose_starter()
    player_pokedex.add_pokemon(starter_pokemon)

    return Team(starter_pokemon, team_name=name)

def steroids(starter_pokemon:Pokemon) -> None:
    while True:
        print('\nWould you like to put your satrter pokemon on steriods?\n')
        answer = input('y/n')
        if answer == 'y':
            starter_pokemon.attack = 800
            return
        elif answer == 'n':
            return
        
def get_player_action() -> str:
    """Gets the player's choice of three actions. returns it"""
    # select action
    action = None
    actions = ['1', '2', '3']
    print('\nWhat will you do?\n1. Battle Pokemon\n2. Change Team\n3. Finish')
    while action not in actions:
        action = input('Choose an action')

        # if you don't have enough pokemon to change you team
        #if action == '2' and len(player_pokedex) <= 6:
        #    continue
    return action

def game_loop(player_team:Team, player_pokedex:Pokedex) -> None:
    """The main game loop that runs actions"""
    winner = player_team
    while winner == player_team:
        action = get_player_action()

        # Pokemon battle:
        if action == '1':
            winner = trigger_battle(player_team, player_pokedex)
        
        # Changing team
        elif action == '2' and len(player_pokedex.collected_pokemon) != len(player_team.full_team):
            change_team(player_team, player_pokedex)
            winner = player_team
            continue

        # Exiting
        elif action == '3':
            break

    if winner != player_team:
        print('Game Over. All your pokemon fainted!')
    
    return

def game_end_summary(player_pokedex:Pokedex) -> None:
    """Prints final messages once game is finished"""
    print(f"\nYou caught {len(player_pokedex)}/{len(ALL_POKEMON)} pokemon! That's {round(len(player_pokedex)/len(ALL_POKEMON) * 100, 2)}%!")

    levels = [pokemon.level for pokemon in player_pokedex.collected_pokemon]
    print(f'You reached level {max(levels)}!')
    return
````
